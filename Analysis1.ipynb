```python
import pandas as pd
```

```python
## open the evictions CSV file
df = pd.read_csv("evictionrates.csv")
df
```

```python
df.info()
```

```python
## count evictions for each borough 
df["BOROUGH"].value_counts()
```

```python
## give me the borough with the highest number of evictions
df.groupby('BOROUGH').size().sort_values(ascending=False).head(1)
```

```python
## find the postal code that has the most evictions 
df["Eviction Postcode"].mode()
```

```python
## list each year with the number of evictions for each borough 
df.groupby(['Year', 'BOROUGH']).size() 
```
